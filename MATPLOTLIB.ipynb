```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
plt.plot(x,y)
plt.show()
#plot:plot is a function used to draw a 2d plot or graph using values on x axis and y axis
```

```python
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
plt.plot(x,y,color='purple')
plt.show()
```

```python
#adding title to the plot and label to axis
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("X vs y plot")
plt.plot(x,y,color='red')
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
a=[1,3,5,7,9,11,13]
b=[2,4,6,8,10,12,14]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("X vs y plot")
plt.plot(x,y,color='red')
plt.plot(a,b,color='green')
plt.show()
```

```python
#displaying legend in the plot
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y,color='red',label='line-1')
a=[3,9,10,4,5]
b=[1,7,3,2,9]
plt.plot(a,b,color='blue',linestyle='--',linewidth =2 ,label='line-2')
plt.legend(loc='best')
plt.show()
```

```python
#markser
x=[1,2,3,4,5,6,7]
y=[2,4,6,8,10,12,14]
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.title("Graph with corner marking")
plt.plot(x,y,color='red',marker='o',markersize=10,markerfacecolor='red',label='line-1')
plt.show()
```

```python
##### write a program to input initail velocity and acceleration and plot the following graph depecting the equation of motion
#v=u+at
#s=ut+1/2at^2
#s=(v^2 -u^2)/2a*s
u=float(input("Enter the initial velocity:"))
a=float(input("Enter the acceleration:"))
t=np.linspace(0, 10, 100)
v= u+a*t
s1=u*t+0.5*a*t**2
s2=(v**2-u**2)/(2*a)
plt.plot(t,v,color='red')
plt.plot(t,s1,color='green')
plt.plot(v,s2,color='blue')
plt.xlabel('Time or Velocity')
plt.ylabel('Velocity / Displacement')
plt.title('Equations of Motion')
plt.show()

```

```python
fig,ax=plt.subplots()
fruits=['apple','blueberry','cherry','orange']
count=[40,100,30,55]
bar_labels=['red','blue','_red','orange']
bar_colors=['tab:red','tab:blue','tab:red','tab:orange']
ax.bar(fruits,count,label=bar_labels,color=bar_colors)
ax.set_ylabel('fruit color')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='fruit color')
plt.show()
```

```python
species=['adelie','chinstrap','gentoo']
sex_count={
    'Male':np.array([73,22,61]),'Female':np.array([73,22,58])}
width=0.6
fig,ax=plt.subplots()
bottom = np.zeros(3)

for sex,sex_count in sex_count.items():
    p=ax.bar(species,sex_count,width,label=sex,bottom=bottom)
    bottom=sex_count
    ax.bar_label(p,label_type='center')
ax.set_title("Pengins by sea")
ax.legend()
plt.show()
```

```python
#piechart: pie() function is used
revenue=[7267,6785,8765,9876,4567]
item=['veg','meat','snacks','drinks','salads']
plt.pie(revenue,labels=item,wedgeprops={'edgecolor':'black'},autopct='%1.1f%%')
plt.title('unique food court')
```

```python
#piechart: pie() function is used
revenue=[7267,6785,8765,9876,4567]
ex=[0,0,1.0,0.1,0]
item=['veg','meat','snacks','drinks','salads']
plt.pie(revenue,labels=item,wedgeprops={'edgecolor':'black'},explode=ex,
        autopct='%1.1f%%',startangle=180)
plt.title('unique food court')
```

```python
#to plot lines lineplot() is used:
sns.lineplot(x='year',y='passengers',data=a)
plt.show()
```

```python
#Bar plots
plt.figure(figsize=(20,10))
sns.set(style='whitegrid')
sns.barplot(x='month',y='passengers',data=a)
plt.show()
```

```python
#Bar plots: Dark Theme
plt.figure(figsize=(20,10))
sns.set(style='dark')
sns.barplot(x='month',y='passengers',data=a)
plt.show()
```

```python
#Bar plots
plt.figure(figsize=(20,10))
sns.set(style='darkgrid')
sns.barplot(x='month',y='passengers',data=a,palette='rocket')
plt.show()
```
